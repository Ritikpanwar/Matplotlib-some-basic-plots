# MATPLOTLIB


```python
#WE start with by importing library
import matplotlib.pyplot as plt
import numpy as np
```

```python
# now plot a line graph
x=[1,3,4,5,2,4,4,6,7,8]
y=[1,2,6,5,6,8,8,9,9,18]
plt.plot(x,y)#plot graph
plt.xlabel("x-axis")# give the name of axis
plt.ylabel("y-axis")# give the name of axis
plt.title("our first graph")
plt.grid()   # use for providing the grid in the back side of graph this is optional   
```

```python
import numpy as np
x=np.arange(0,2,0.01)
y=np.cos(2*np.pi*x)
plt.plot(x,y)#plot graph
plt.xlabel("x-axis")# give the name of axis
plt.ylabel("y-axis")# give the name of axis
plt.title("our first graph")
# use for providing the grid in the back side of graph this is optional , as you can see we not use in this graph
```

## Subplot

```python
x=[1,3,4,5,2,4,4,6,7,8]
y=[1,2,6,5,6,8,8,9,9,18]

plt.subplot(2,1,1)# by this we divide a plot in two parts and work on it  


plt.plot(x,y,"m")# m giving the color of the graph line 
plt.xlabel("x-axis")# give the name of axis
plt.ylabel("y-axis")# give the name of axis
plt.title("subplot 1")


plt.subplot(2,1,2)
x=np.arange(0,2,0.01)
y=np.cos(2*np.pi*x)
plt.plot(x,y,"g","--")# design of the line is givin by the "--"
plt.xlabel("x-axis")# give the name of axis
plt.ylabel("y-axis")# give the name of axis
plt.title("subplot 2")
# use for providing the grid in the back side of graph this is optional , as you can see we not use in this graph
```

## BARPLOT

```python
x=[3,5,7,2,7,7,8]
y=[54,57,86,46,43,99,36]
tick_label=[2,3,4,5,6,7,8]
plt.bar(x,y,tick_label=tick_label,color=['blue','green'],label="gf")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()# if we not use the then label is not show 
plt.title("barplot")
plt.show()
```

## Histogram

```python
x=[40,35,7,57,36,47,7,47,57,53,95,36,8,36,64,73,74,83,93,39,64,35,47,64,40,45,56]
bin=10
rang=(0,100)
plt.hist(x,bin,rang,rwidth=0.7,color='green',label="ages")# rwidth is the width btw the two bars of of histogram 
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.title("histogram")
plt.show()
```

## scatter plot 

```python
x=['a','b','c','d','e','f','g','h','i','j']
y=[3,4,6,2,4,7,9,4,4,3]
plt.scatter(x,y,label='star',marker='*',s=100)#s is the size of marker
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scaatterplot")
plt.show()
```

```python
y=['hollywood','south','tamil','bollywood']
x=[7000,5000,2000,10000]
colors=['g','r','m','b']

plt.pie(x,labels=y,colors=colors,startangle=90,shadow=True,explode=(.2,0,0,0),autopct="%1.2f%%")
plt.legend()
```

### Read this for pie chart

```python
#explode have 4 value because chart has for part if you change third value of explode the 3 part will sepearte
# start angle give the angle on which pie chart start
# autopct give percentage
```
